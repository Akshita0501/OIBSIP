```python
import zipfile
import os

# Path to the zip file in Google Drive
zip_path = '/content/drive/MyDrive/unemployement.zip'


```

```python
# Directory to extract the files
extracted_path = '/content/unemployement_dataset'

# Create the directory if it doesn't exist
os.makedirs(extracted_path, exist_ok=True)

# Extract the contents of the zip file
with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extracted_path)
```

Load and Analyse the data:

```python
import pandas as pd

# Load the datasets
india_df = pd.read_csv(os.path.join(extracted_path, 'Unemployment in India.csv'))
rate_df = pd.read_csv(os.path.join(extracted_path, 'Unemployment_Rate_upto_11_2020.csv'))

# Display the first few rows of the datasets
print("Unemployement in India dataset:")
print(india_df.head())

print("\nUnemployement rate up to Nov 2020 dataset:")
print(rate_df.head())
```

```python
print(rate_df.columns)
```

Visualize the data:

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Remove leading space from the ' Date' column name
rate_df[' Date'] = pd.to_datetime(rate_df[' Date'])

# Assuming you have a ' Date' and ' Estimated Unemployment Rate (%)' column in your datasets
plt.figure(figsize=(12, 6))
sns.lineplot(x=' Date', y=' Estimated Unemployment Rate (%)', data=rate_df)
plt.title('Estimated Unemployment Rate Over Time')
plt.xlabel('Date')
plt.ylabel('Estimated Unemployment Rate (%)')
plt.xticks(rotation=45)
plt.show()
```

Descriptive Statistics:

```python
# Summary statistics
print(rate_df.describe())

# Histogram of the estimated unemployment rate
plt.figure(figsize=(10, 6))
sns.histplot(rate_df[' Estimated Unemployment Rate (%)'], bins=20, kde=True)
plt.title('Distribution of Estimated Unemployment Rate')
plt.xlabel('Estimated Unemployment Rate (%)')
plt.ylabel('Frequency')
plt.show()
```

Regional Analysis:

```python
# Boxplot of unemployment rates by region
plt.figure(figsize=(14, 8))
sns.boxplot(x='Region', y=' Estimated Unemployment Rate (%)', data=rate_df)
plt.title('Unemployment Rate Variation Across Regions')
plt.xlabel('Region')
plt.ylabel('Estimated Unemployment Rate (%)')
plt.xticks(rotation=45)
plt.show()
```
